```python
import pandas as pd
import numpy as np

features = pd.read_csv("female.csv")
target = pd.read_csv("id.csv")
features1 = pd.DataFrame(features)
target1 = pd.DataFrame(target)
target1
features1
```

```python
method1 = pd.merge(features1, target1,how="inner")
display(method1)
print("Length:",len(method1))
```

```python
method2 = pd.merge(features1, target1,how="outer")
display(method2)
print("Length:",len(method2))
```

```python
method4 = pd.merge(features1,target1,on = "id",how ="right")
display(method4)
print("Length:",len(method4))
```

```python
method4 = pd.merge(features1,target1,on="id",how ="left")
display(method4)
print("Length:",len(method4))
```
