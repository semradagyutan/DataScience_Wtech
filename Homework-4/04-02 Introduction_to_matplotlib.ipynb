1.Create a new figure with matplotlib functions. Plot cosine function for the numbers between -3 π  and 3 π  with 0.1 intervals.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x= 3*(np.pi)
y= 3*(-np.pi)
z = np.arange(y,x, 0.1)
k = np.cos(z)

fig, ax = plt.subplots() 
ax.plot(z,k,color = "gray")
plt.title("cos")
plt.show()
```
