1-First, filter the data for four countries (Spain, France, Germany, Italy) and the last 100 days.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
corona = pd.read_csv('owid-covid-data.csv',index_col=2)

g = corona.loc["Germany"].tail(100)
s = corona.loc["Spain"].tail(100)
i = corona.loc["Italy"].tail(100)
f = corona.loc["France"].tail(100)

four_countries = pd.concat([g,s,i,f])


display(four_countries)
```

2-Create a figure with two subplots. The left subplot will show total cases, and the right one will show total deaths for these countries

3-Set titles, X and Y labels, change the font, font size, and font color for each subplot.

4-Change color, width, and type of lines.

5-Put a legend to bottom right.

6-Change the color and rotation of the X-axis ticks.

7-And set a title for the figure (not for the plots).

```python
plt.figure(figsize=(20,6))

plt.subplot(1,2,1)
plt.plot(four_countries["total_cases"], c = "red", markersize = 10, linestyle = "-",linewidth = 3)
plt.xticks(rotation = 0 ,fontsize = 20)
plt.title("Total Cases",fontsize = 20, color = "red")

plt.xlabel("Countries",fontstyle = "normal",fontsize=15,fontweight = "normal", color = "blue")
plt.ylabel("Total cases",fontstyle = "normal",fontsize=15,fontweight = "normal", color = "blue")
plt.legend("cases",loc=4)

plt.subplot(1,2,2)
plt.plot(four_countries["total_deaths"],c = "green", markersize = 10, linestyle = "-",linewidth = 3)
plt.xticks(rotation = 0, fontsize = 20)
plt.title("Total Death",fontsize = 20, color = "green")


plt.xlabel("Countries",fontstyle = "oblique",fontsize=15,fontweight = "normal", color = "purple")
plt.ylabel("Total Deaths",fontstyle = "oblique",fontsize=15,fontweight = "normal", color = "purple")
plt.legend("death",loc=4)

plt.show()
```
