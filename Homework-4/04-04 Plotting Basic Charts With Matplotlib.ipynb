In this assignment, you will continue work with the Coronavirus Source Data. You will plot different chart types. Don't forget to set titles and axis labels.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


corona = pd.read_csv("owid-covid-data.csv")
total_case = corona.groupby(by="location")["new_cases"].sum()
sorting = (total_case.sort_values(ascending= False)).head(21)

high20countries = sorting.drop("World", axis=0)
pd.Series(high20countries)                            
                             
cities = []                             
for city in high20countries.index[0:21]:
    cities.append(city)

print(cities)
print(high20countries)
```

Plot a bar chart for total cases of the 20 countries that have the highest numbers.

```python
plt.figure(figsize=(10, 5))
plt.title('Total Cases of the 20 Countries')
plt.xlabel("Countries")
plt.ylabel("Total Cases")
plt.bar(cities, high20countries, color = "blue")
plt.xticks(rotation = 90, fontsize = 11)
plt.show()
```

Plot a histogram for daily deaths for any country you choose. Make three subplots for different bins.

```python
import warnings
warnings.filterwarnings('ignore')

corona = pd.read_csv("owid-covid-data.csv",index_col=0)
title_style_small = {'family': 'Century Gothic', 'color': 'green', 'size': 10 }
axis_style_small  = {'family': 'Century Gothic', 'color': 'blue', 'size': 10 }

plt.figure(figsize=(15, 5))

plt.subplot(1,3,1)
plt.title("New deaths", fontdict = title_style_small )
for city in ["Turkey"]:
    plt.hist(corona[corona.location == city].new_deaths,color = "yellow",bins = 50)
    plt.xlabel("Daily deaths", fontdict = axis_style_small)

plt.subplot(1,3,2)
plt.title("New deaths", fontdict = title_style_small )
for city in ["Turkey"]:
    plt.hist(corona[corona.location == city].new_deaths,color = "purple",bins = 30)
    plt.xlabel("Daily deaths", fontdict = axis_style_small)

plt.subplot(1,3,3)
plt.title("New deaths", fontdict = title_style_small )
for city in ["Turkey"]:
    plt.hist(corona[corona.location == city].new_deaths,color = "green",bins = 20)
    plt.xlabel("Daily deaths", fontdict = axis_style_small)
```

Plot a scatter plot of new cases and new death for Germany and France.

```python
corona = pd.read_csv("owid-covid-data.csv",index_col=0)

plt.figure(figsize=(10, 6))
plt.title(' New Cases and New Death for Germany and France')
for city in ['Germany', 'France']:
    plt.scatter(corona[corona.location == city].new_deaths, 
                corona[corona.location == city].new_cases, label = city)
plt.xlabel('New Deaths')
plt.ylabel('New Cases')
plt.legend()
plt.show()
```

Plot a boxplot for daily deaths for any country you choose.


```python
corona = pd.read_csv("owid-covid-data.csv",index_col=0)
plt.figure(figsize=(10, 6))

title_style = {'family': 'Century Gothic', 'color': 'green', 'size': 15 }
axis_style  = {'family': 'Century Gothic', 'color': 'blue', 'size': 15 }


plt.title('Daily deaths for Italy', fontdict = title_style)
plt.boxplot(corona[(corona['location']=='Italy')]["new_deaths"])
plt.xlabel('New Deaths',fontdict = axis_style)
plt.ylabel('count', fontdict = axis_style)
plt.show()
```

Calculate the total case for each continent and plot a pie chart.

```python
corona = pd.read_csv("owid-covid-data.csv",index_col=0)

total_cases = corona.groupby(by="continent")["new_cases"].sum()
print(total_cases)

continents = []                             
for continent in total_cases.index:
    continents.append(continent)

print(continents)

plt.figure(figsize=(10, 6))
plt.title('Total case for each Continent', fontdict = title_style)

plt.pie(total_cases,labels=continents, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.show()
```
