Consider how you can fill in missing values for each variable. Filling in the missing values with a value is meaningful for which variables and for which ones?

```python
states["ENROLL"].fillna(states["ENROLL"].mean(), inplace=True)
states.isnull().sum()*100/states.shape[0]
states.ENROLL.mean()
```

3-Now it's time to consider the factor of time! Review your answer in question 2 and fill in the missing data based on the values observed during that year. For example, if you want to fill a value with an average value, calculate the average of that year.

```python
states = pd.read_csv("datasets_95317_1078789_states_all.csv")

for i in range(0,len(states)):
    if pd.isnull(states["ENROLL"][i])==True:
        states["ENROLL"][i]= states.groupby("YEAR").mean()["ENROLL"][states["YEAR"][i]]
        
print(  states["ENROLL"].mean())
```
