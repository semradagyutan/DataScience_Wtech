```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from  scipy.stats import zscore
import seaborn as sns
from sklearn import linear_model
import statsmodels.api as sm
import warnings
warnings.filterwarnings('ignore')
pd.options.display.max_columns = None
```

```python
weather = pd.read_csv("weatherHistory.csv")
weather.head()
```

```python
print(weather.isnull().sum())
print("**"*22)
print(weather.isnull().sum()*100/weather.shape[0])
```

```python
weather.info()
```

```python
#Target Variable
weather["TempatureDif"] = weather["Temperature (C)"] - weather["Apparent Temperature (C)"]

Y = weather["TempatureDif"]

X = weather[["Humidity","Wind Speed (km/h)"]]
X = sm.add_constant(X)

results = sm.OLS(Y,X).fit()
display(results.summary())
```

```python
house_train.drop(['PoolQC', "Fence", "MiscFeature","FireplaceQu","Alley","LotFrontage"],axis=1,inplace=True)
```

```python
print(house_train.isnull().sum()*100/house_train.shape[0])
```

```python
house_train["GarageYrBlt"].fillna(house_train["GarageYrBlt"].mode()[0],inplace=True)
house_train["GarageCond"].fillna(house_train["GarageCond"].mode()[0],inplace=True)
house_train["GarageFinish"].fillna(house_train["GarageFinish"].mode()[0],inplace=True)
house_train["GarageArea"].fillna(house_train["GarageArea"].mode()[0],inplace=True)
house_train["GarageQual"].fillna(house_train["GarageQual"].mode()[0],inplace=True)
house_train["GarageCars"].fillna(house_train["GarageCars"].mode()[0],inplace=True)
house_train["GarageType"].fillna(house_train["GarageType"].mode()[0],inplace=True)

house_train.dropna(inplace=True)
```

Öncelikle hangileri ile ilgili olabilir diye yaptığım incelemeden sonra --> Steet,GarageQual,GarageCars,CentralAir,HeatingQC, TotRmsAbvGrd,GarageArea

```python
house_train["central_air"]=pd.get_dummies(house_train["CentralAir"], drop_first=True)

house_train["Street_"]=pd.get_dummies(house_train["Street"], drop_first=True)
```

```python
def heatingquc(HeatingQC):
    if HeatingQC=="Ex":
        return 5
    elif HeatingQC=="Gd":
        return 4
    elif HeatingQC=="TA":
        return 3
    elif HeatingQC=="Fa":
        return 2
    elif HeatingQC=="Po":
        return 1


house_train["HeatingQc"] = house_train["HeatingQC"].apply(heatingquc)    
```

```python
def Garagequal(GarageQual):
    if GarageQual=="Ex":
        return 6
    elif GarageQual=="Gd":
        return 5
    elif GarageQual=="TA":
        return 4
    elif GarageQual=="Fa":
        return 3
    elif GarageQual=="Po":
        return 2
    elif GarageQual=="NA":
        return 1


house_train["GarageQuall"] = house_train["GarageQual"].apply(Garagequal)    
```

```python
def kitchenqual(KitchenQual):
    if KitchenQual=="Ex":
        return 4
    elif KitchenQual=="Gd":
        return 3
    elif KitchenQual=="TA":
        return 2
    elif KitchenQual=="Fa":
        return 1

house_train["Kitchen_Qual"]=house_train["KitchenQual"].apply(kitchenqual)
```

```python
Y_ = house_train["SalePrice"]

X_ = house_train[["HeatingQc","GarageQuall","GarageCars",
                 "central_air","TotRmsAbvGrd","GarageArea","Kitchen_Qual","GrLivArea","Street_"]]
X_ = sm.add_constant(X_)

results = sm.OLS(Y_,X_).fit()
display(results.summary())
```

Sonuçlara göre;

GarageQuall: Garaj kalitesindeki bir birimlik artış 1.731e+04 fiyat üzerinde artma görülmektedir.

GarageCars: Garajın aldığı araba sayısındaki bir birimlik artış 2.134e+04 fiyat üzerinde artma görülmektedir.

central_air: Merkezi klimanın olması 1.979e+04 kadar etkilemektedir.

TotRmsAbvGrd: Oda sayısındaki bir birimlik artış fiyatı 1.537e+04 etkilemektedir.

GarageArea: Garaj alanındaki bir birimlik artış fiyatı 49.4966 arttırmaktadır.

Kitchen_Qual: Kalitedeki bir birimlik iyileşme fiyatı 4.458e+04	arttırmaktadır.
