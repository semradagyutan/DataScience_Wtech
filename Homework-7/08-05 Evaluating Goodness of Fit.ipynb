```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from  scipy.stats import zscore
import seaborn as sns
from sklearn import linear_model
import statsmodels.api as sm
import warnings
warnings.filterwarnings('ignore')
pd.options.display.max_columns = None
```

```python
weather = pd.read_csv("weatherHistory.csv")
weather.head()
```

```python
#Target Variable
weather["TempatureDif"] = weather["Temperature (C)"] - weather["Apparent Temperature (C)"]

Y = weather["TempatureDif"]

X = weather[["Humidity","Wind Speed (km/h)"]]
X = sm.add_constant(X)

results = sm.OLS(Y,X).fit()
display(results.summary())
```

Sonuçlara göre R-squared ve Adj. R-squared değerleri yüzde 28 değer aldığı görülüyor. Modelin başarısı yani hedef değişkeni açıklama oranı çok düşük. Bu yüzden tatmin edici olmadığı söylenebilir. 

```python
#Target Variable
weather["Hum/Wind"] = weather["Humidity"]*weather["Wind Speed (km/h)"]
Y = weather["TempatureDif"]

X = weather[["Humidity","Wind Speed (km/h)","Hum/Wind"]]
X = sm.add_constant(X)


results = sm.OLS(Y,X).fit()
display(results.summary())
```

Önceki modele göre ilerleme vardır. Çünkü R-squared değerler %34.1 olmuştur. Bu da yeni eklediğimiz feature'ın modelin başarısını arttırdığını söyleyebiliriz.


```python
Y = weather["TempatureDif"]

X = weather[["Humidity","Wind Speed (km/h)","Visibility (km)"]]
X = sm.add_constant(X)

results = sm.OLS(Y,X).fit()
display(results.summary())
```

İlk modele eklediğimiz visibility özelliği sayesinde ilk modele göre r-squared değerleri artmıştır. Fakat açıklama oranları kıyaslaması açısından önceki modele eklediğimiz interaction bu modele göre daha fazla açıklama başarısı elde etmiştir.

```python
house_train = pd.read_csv("house_train.csv")
house_train.head()
```

```python
house_train.drop(['PoolQC', "Fence", "MiscFeature","FireplaceQu","Alley","LotFrontage"],axis=1,inplace=True)
```

```python
house_train["GarageYrBlt"].fillna(house_train["GarageYrBlt"].mode()[0],inplace=True)
house_train["GarageCond"].fillna(house_train["GarageCond"].mode()[0],inplace=True)
house_train["GarageFinish"].fillna(house_train["GarageFinish"].mode()[0],inplace=True)
house_train["GarageArea"].fillna(house_train["GarageArea"].mode()[0],inplace=True)
house_train["GarageQual"].fillna(house_train["GarageQual"].mode()[0],inplace=True)
house_train["GarageCars"].fillna(house_train["GarageCars"].mode()[0],inplace=True)
house_train["GarageType"].fillna(house_train["GarageType"].mode()[0],inplace=True)

house_train.dropna(inplace=True)
```

```python
house_train["central_air"]=pd.get_dummies(house_train["CentralAir"], drop_first=True)
```

```python
def heatingquc(HeatingQC):
    if HeatingQC=="Ex":
        return 5
    elif HeatingQC=="Gd":
        return 4
    elif HeatingQC=="TA":
        return 3
    elif HeatingQC=="Fa":
        return 2
    elif HeatingQC=="Po":
        return 1


house_train["HeatingQc"] = house_train["HeatingQC"].apply(heatingquc)  
```

```python
def Garagequal(GarageQual):
    if GarageQual=="Ex":
        return 6
    elif GarageQual=="Gd":
        return 5
    elif GarageQual=="TA":
        return 4
    elif GarageQual=="Fa":
        return 3
    elif GarageQual=="Po":
        return 2
    elif GarageQual=="NA":
        return 1


house_train["GarageQuall"] = house_train["GarageQual"].apply(Garagequal)  
```

```python
def kitchenqual(KitchenQual):
    if KitchenQual=="Ex":
        return 4
    elif KitchenQual=="Gd":
        return 3
    elif KitchenQual=="TA":
        return 2
    elif KitchenQual=="Fa":
        return 1

house_train["Kitchen_Qual"]=house_train["KitchenQual"].apply(kitchenqual)
```

```python
Y_ = house_train["SalePrice"]

X_ = house_train[["HeatingQc","GarageQuall","GarageCars",
                 "central_air","TotRmsAbvGrd","GarageArea","Kitchen_Qual","GrLivArea","LotArea"]]
X_ = sm.add_constant(X_)

results = sm.OLS(Y_,X_).fit()
display(results.summary())
```

Değerlere bakıldığı zaman 

R-squared = %72.3 Modelin başarısı arttırılmadılır. 
F-Statistic = 0. Modelin anlamlı olduğunu söylemektedir.


Modelin başarısı daha iyileştirilebilir olduğu için tam olarak tatmin edici olduğu söylenemez. 
Bu yüzden anlamlı olmayan feature'ları çıkarıp anlamlı olan feature'lar eklenmelidir.
